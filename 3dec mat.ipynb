```python
avg_price_room_type =df.groupby("room type")['price'].mean()
plt.barh(avg_price_room_type.index, avg_price_room_type.values, color='orange')
plt.title("Average Price by Room Type")
plt.xlabel("Average Price ($)")
plt.ylabel("Room Type")
plt.show()
```

```python
plt.hist(df['price'], bins=50, color='blue', alpha=0.7)
plt.title("Price Distribution")
plt.xlabel("Price ($)")
plt.ylabel("Frequency")
plt.yscale('log')  # Optional log scale
plt.show()
```
